```python
# Import necessary packages, set up Latex font
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot parameters
Delta = 1     # Fixing Delta=1 or simplicity
N = 500
x_min=-0.5
x_max = 7.5
x_range = np.linspace(x_min, x_max, N)
n=3
```

```python
# Make plot
fig, ax = plt.subplots(n,1, figsize=(8, 5), dpi=200)
for k in range(1,n+1):
  function_vals = np.heaviside(np.cos(np.pi/2**(n-k)*(x_range - 2**(n-k-1) + 1/2)), 0)
  ax[k-1].plot(x_range, function_vals, label='$S_k(x)=1-x_k$')
  ax[k-1].legend()
  ax[k-1].set_xlabel('$x$', fontsize=18)
  ax[k-1].set_xlim(x_min,x_max)
  ax[k-1].set_yticks([0,1])
  ax[k-1].xaxis.set_tick_params(labelsize=12)
  ax[k-1].yaxis.set_tick_params(labelsize=12)
  ax[k-1].legend(prop={'size': 12})
```

```python
# Make plot with thicker lines
fig, ax = plt.subplots(n,1, figsize=(8,4.7), dpi=200)

k=1
function_vals = np.heaviside(np.cos(np.pi/2**(n-k)*(x_range - 2**(n-k-1) + 1/2)), 0)
ax[k-1].plot(x_range, function_vals, label='$S_1(\hat{x})=1-x_1$', linewidth=3)
ax[k-1].legend()
#ax[k-1].set_xlabel('$\hat{x}$', fontsize=18)
ax[k-1].set_xlim(x_min,x_max)
ax[k-1].set_yticks([0,1])
ax[k-1].xaxis.set_tick_params(labelsize=14)
ax[k-1].yaxis.set_tick_params(labelsize=14)
ax[k-1].legend(prop={'size': 14}, loc='upper right', bbox_to_anchor=(0.97, .95))


k=2
function_vals = np.heaviside(np.cos(np.pi/2**(n-k)*(x_range - 2**(n-k-1) + 1/2)), 0)
ax[k-1].plot(x_range, function_vals, label='$S_2(\hat{x})=1-x_2$', linewidth=3)
ax[k-1].legend()
#ax[k-1].set_xlabel('$\hat{x}$', fontsize=18)
ax[k-1].set_xlim(x_min,x_max)
ax[k-1].set_yticks([0,1])
ax[k-1].xaxis.set_tick_params(labelsize=14)
ax[k-1].yaxis.set_tick_params(labelsize=14)
ax[k-1].legend(prop={'size': 14}, loc='upper right', bbox_to_anchor=(0.97, .95))


k=3
function_vals = np.heaviside(np.cos(np.pi/2**(n-k)*(x_range - 2**(n-k-1) + 1/2)), 0)
ax[k-1].plot(x_range, function_vals, label='$S_3(\hat{x})=1-x_3$', linewidth=3)
ax[k-1].legend()
ax[k-1].set_xlabel('$\hat{x}$', fontsize=18)
ax[k-1].set_xlim(x_min,x_max)
ax[k-1].set_yticks([0,1])
ax[k-1].xaxis.set_tick_params(labelsize=14)
ax[k-1].yaxis.set_tick_params(labelsize=14)
ax[k-1].legend(prop={'size': 14}, loc='upper right', bbox_to_anchor=(0.97, .95))


fig.tight_layout()
```
